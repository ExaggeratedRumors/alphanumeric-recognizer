```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Read data

```python
train_set = pd.read_csv("emnist-balanced-train.csv")
test_set = pd.read_csv("emnist-balanced-test.csv")

train_set.head()
```

## About training data

```python
print(f"Info:")
train_set.info()
print(f"\nShape:\n{train_set.shape}")



labels_map = "emnist-balanced-mapping.txt"
mapping = {}
with open(labels_map, "r") as f:
    for line in f:
        key, value = map(int, line.split())
        mapping[key] = value

labels_names = train_set["45"].values
plt.figure(figsize=(20,6))
sns.countplot(x=labels_names, palette="hls", legend=False)
translated_labels = [chr(mapping[label]) for label in np.unique(labels_names)]
ax = plt.gca()
ax.set_xticklabels(translated_labels)
plt.show()
```

## Visualization

```python
x_train = np.array(train_set.iloc[:,1:].values)
y_train = np.array(train_set.iloc[:,0].values)
fig, axes = plt.subplots(6, 6,figsize=(10,6))
for i, j in enumerate(axes.flat):
    j.imshow(x_train[i + 2].reshape([28,28]))
    j.set_title(chr(mapping[y_train[i + 2]]))
fig.tight_layout()
```
